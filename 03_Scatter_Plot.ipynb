# Scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd


```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.scatter(x,y)
plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
plt.scatter(x,y)

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
plt.scatter(x,y , color = "r")

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
colors = ["r" , "b" , "g" ,"m" , "r" , "g" , "w" , "r" , "g"]
plt.scatter(x,y , c = colors)

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
sizes = [100 , 200 , 300, 400, 500,600,700,800,900]

colors = ["r" , "b" , "g" ,"m" , "r" , "g" , "w" , "r" , "g"]
plt.scatter(x,y , c = colors , s= sizes)

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
sizes = [100 , 200 , 300, 400, 500,600,700,800,900]

colors = ["r" , "b" , "g" ,"m" , "r" , "g" , "w" , "r" , "g"]
plt.scatter(x,y , c = colors , s= sizes , alpha = 0.5)

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
sizes = [100 , 200 , 300, 400, 500,600,700,800,900]

colors = ["r" , "b" , "g" ,"m" , "r" , "g" , "w" , "r" , "g"]
plt.scatter(x,y , c = colors , s= sizes , alpha = 1 , marker = "*")

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
sizes = [100 , 200 , 300, 400, 500,600,700,800,900]

colors = ["r" , "b" , "g" ,"m" , "r" , "g" , "w" , "r" , "g"]
plt.scatter(x,y , c = colors , s= sizes , alpha = 1 , marker = "*" , edgecolor = "g" , linewidth = 2)

plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
sizes = [100 , 200 , 300, 400, 500,600,700,800,900]

colors = [10 , 49, 44 , 52 , 78, 30 , 20, 65,89]

plt.scatter(x,y , c = colors , s= sizes , cmap = "Blues")
t = plt.colorbar()
t.set_label("color bar", fontsize = 16)
plt.show()
```

```python
x = [2,4,6,8,10,2,4,2,6]
y = [1,2,3,4,3,6,8,8,9]
z = [11,13,2,1,3,2,9,6,5]

plt.title("Scatter plot" , fontsize = 15)
plt.xlabel("DAY")
plt.ylabel("NO.")
sizes = [100 , 200 , 300, 400, 500,600,700,800,900]

colors = [10 , 49, 44 , 52 , 78, 30 , 20, 65,89]
plt.scatter(x,z , color = "r" , s = sizes)
plt.scatter(x,y , c = colors , s= sizes , cmap = "Blues")
plt.colorbar()

plt.show()
```
