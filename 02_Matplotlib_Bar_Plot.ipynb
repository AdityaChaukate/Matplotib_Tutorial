```python
x = ['Python' , 'C++' , 'C' , 'Java']
y = [85,70,60,82]
z = [20 , 30 ,40 , 50]

plt.title("XYZ")
plt.xlabel("Languages" , fontsize = 10)
plt.ylabel("Popularity")
c = ['r']
plt.bar(x,y,width = 0.5 , color = c , label = "Popularity")
plt.bar(x,z,width = 0.5 , label = "ABC")

plt.legend()
plt.show()
```

```python
x = ['Python','C++','C','Java']
y = [85,70,60,82]
z = [20 , 30 ,40 , 50]

width = 0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Languages" , fontsize = 10)
plt.ylabel("Popularity")
plt.title("XYZ" , fontsize = 15)

plt.bar(p,y,width, color = "r" , label = "Popularity")
plt.bar(p1,z,width, label = "ABC")


plt.xticks(p+width / 2, x ,rotation = 90)
plt.legend()
plt.show()
```

```python
x = ['Python' , 'C++' , 'C' , 'Java']
y = [85,70,60,82]
z = [20 , 30 ,40 , 50]

plt.title("XYZ")
plt.xlabel("Languages" , fontsize = 10)
plt.ylabel("Popularity")
c = ['r']
plt.barh(x,y, color = c , label = "Popularity")
plt.barh(x,z, label = "ABC")

plt.legend()
plt.show()
```
