```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1)   # 1 row , 2 column , 1st index
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)  # 1 row , 2 column , 2nd index
plt.bar(x,y)
plt.title("Bar Chart")

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # Unpack correctly

ax[0].plot(x, y, color='blue')
ax[0].set_title("Line Plot")

ax[1].bar(x, y, color='green')
ax[1].set_title("Bar Plot")

plt.show()
```
